```python
# Only need to run once to preprocess the metadata
# Original metadata from ReDU database

redu_metadata = pd.read_csv("data/all_sampleinformation.tsv", sep="\t", dtype=str)
SKIN_CATEGORY = "skin of trunk|skin of pes|head or neck skin|axilla skin|skin of manus|arm skin|skin of leg"
BLOOD_CATEORY = "blood plasma|blood serum"

redu_metadata["USI"] = redu_metadata["USI"].apply(lambda x: process_USI(x))
redu_metadata["UBERONBodyPartName"] = redu_metadata["UBERONBodyPartName"].replace(to_replace=SKIN_CATEGORY, value="skin")
redu_metadata["UBERONBodyPartName"] = redu_metadata["UBERONBodyPartName"].replace(to_replace=BLOOD_CATEORY, value="blood")
redu_metadata["HealthStatus"] = redu_metadata["HealthStatus"].apply(lambda x : x.lower())
redu_metadata.to_csv("data/all_sampleinformation_redu_preprocessed.csv", index=False)
```

```python
REDU_USEFUL_COLUMNS = ['SampleType',
       'SampleTypeSub1',
       'YearOfAnalysis', 'UBERONBodyPartName', 'BiologicalSex', 'AgeInYears',
       'LifeStage', 'Country', 'HealthStatus', 'ChromatographyAndPhase',
       'IonizationSourceAndPolarity', 'MassSpectrometer',
       'SampleExtractionMethod', 'SampleCollectionMethod',
       'ComorbidityListDOIDIndex', 'DOIDCommonName', 'DOIDOntologyIndex',
       'ENVOEnvironmentBiome', 'DepthorAltitudeMeters',
       'HumanPopulationDensity', 'InternalStandardsUsed',
       'LatitudeandLongitude', 'SampleCollectionDateandTime',
       'ENVOEnvironmentMaterial', 'ENVOEnvironmentBiomeIndex',
       'ENVOEnvironmentMaterialIndex', 'SubjectIdentifierAsRecorded',
       'TermsofPosition', 'UBERONOntologyIndex', 'UniqueSubjectID',
       'DataSource']
redu_metadata_useful = redu_metadata[REDU_USEFUL_COLUMNS]
```

```python
# Get unique column values and store all in a single csv
unique_values = {}
for column in redu_metadata_useful.columns:
    unique_values[column] = redu_metadata_useful[column].dropna().unique()

unique_values_df = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in unique_values.items()]))
unique_values_df.to_csv("data/unique_values_redu.csv", index=False)    
```
